```python
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
```

```python
CATEGORIES=['Covid19 Negative','Covid19 Positive']
```

```python
DIRECTORY=r'E:\aine\project9\Project9\DataSet'
```

## Image Preprocessing

```python
img_size=100
for category in CATEGORIES:
    path=os.path.join(DIRECTORY,category)
    for img in os.listdir(path):
        img_path=os.path.join(path,img)
        label=CATEGORIES.index(category)
        arr=cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
        new_arr=cv2.resize(arr,(img_size,img_size))
        plt.imshow(new_arr)
        break
```

```python
data=[]
img_size=100
for category in CATEGORIES:
    path=os.path.join(DIRECTORY,category) # path of the dataset
    for img in os.listdir(path):
        try:
            img_path=os.path.join(path,img) # image path
            label=CATEGORIES.index(category) #assiging label to image; Negative-0; positive-1
            arr=cv2.imread(img_path,cv2.IMREAD_GRAYSCALE) #converting image to Gray scale
            new_arr=cv2.resize(arr,(img_size,img_size))
            data.append([new_arr,label])
        except Exception as e:
            print(str(e))
```

```python
X=[]
y=[]
```

```python
for features,label in data:
    X.append(features) #Storing Images all images in X
    y.append(label) #Storing all image labels in Y
```

```python
X=np.array(X)
y=np.array(y)
```
